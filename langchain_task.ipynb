```python
#Main Function
def genrate_table_summary(filepath: str, selected_prompt : int = 3, m_temperature : float = 0.9):
  """
  Generate a summary for the given table.
  Using Langchain
    Args:
        filepath (str): The path to the file containing the table details.
        selected_prompt (int): The selected prompt option for generating the summary.

    Returns:
        str: The generated table summary.

    Raises:
        FileNotFoundError: If the specified file path does not exist.
  """
  try:
    import pandas as pd
    from langchain.llms import OpenAI

    # LangChain object
    llm = OpenAI(openai_api_key=openAIkey,temperature=m_temperature)
    df = pd.read_csv(filepath)  # Initializing the dataframe
    # collecting columns and their datatypes
    column_names = list(df.columns)
    column_dtypes = [str(data_type_str) for data_type_str in list(df.dtypes)]

    # create table infomational string used in propmt
    tableInfo = "  ".join([f"* Column name {item[0]} having data type {item[1]}\n" for item in zip(column_names,column_dtypes)])

    if selected_prompt == 1:
      prompt = f"""
      Generate a summary for the given table\n{tableInfo}
      Make the summary effective and small.
      """
    elif selected_prompt == 2:
      prompt = f"""
      Given the details for the following table\n{tableInfo}
      Description:
      Generate summary of a table that must define purpose of this table.
      """
    elif selected_prompt == 3:
      prompt = f"""
      Given a details for the table\n{tableInfo}
      Description:
      Generate a summary of a table that must define the purpose of this table.
      Make the summary effective and small.
      """
    elif selected_prompt == 4:
      prompt = f"""
      **Generate a summary for the given table**
      **Table Columns:**
      {tableInfo}
      """

    print("[INFO] Prompt created by function",prompt)

    # pass text and get predictions!
    summary = llm.predict(prompt)
    return  summary
  except Exception as e:
    return  f"[ERR] Error Found | {e}"
```

```python
# Run function
result = genrate_table_summary("in/tables/deals.csv",3)
print("\n","**"*30)
print("Summary for the given table:\n",result)
print("\n","**"*30)
```
